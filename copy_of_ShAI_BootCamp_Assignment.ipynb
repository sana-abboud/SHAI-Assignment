# Check missing values

```python
# Check for missing values in each column
print("Missing Values:\n", df.isnull().sum())
```

```python
# Handle missing values by dropping columns with missing data in the 'Notes,Status' columns
df.drop('Notes',axis=1,inplace=True)
df.drop('Status',axis=1,inplace=True)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# Handle missing values by dropping rows with missing data in the 'BasePay' column
df.dropna(subset=['BasePay'], inplace=True)
```

```python
df.isnull().sum()
```

```python
# Handle missing values by median in the 'Benefits' column
median_Benefits = df['Benefits'].median()
df['Benefits'].fillna(median_Benefits, inplace=True)
```

```python
df.isnull().sum()
```

# Check duplicated

```python
df.duplicated().sum()
```

```python
# Calculate the mean, std, minimum, and maximum salary
print(df['TotalPay'].describe())
```

```python
# Determine the range of salaries
print(df['TotalPay'].max() - df['TotalPay'].min())
```

```python
# Find the mode of salaries
print(df['TotalPay'].mode())
```

```python
# Find the median of salaries
print(df['TotalPay'].median())
```

# Plots

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Splitting categorical and numerical columns
categorical_columns = df.select_dtypes(include=['object']).columns
numerical_columns = df.select_dtypes(include=['float']).columns
```

```python
for i in categorical_columns:
  print(i)
  print(df[i].unique())
  print(df[i].nunique())
  print()
  print(df[i].value_counts())
  print()
```

```python
df.hist(figsize=(12, 10))
plt.show()
```

```python
from scipy.stats import skew
from scipy.stats import skew, boxcox
print(df[numerical_columns].columns)
print(skew(df[numerical_columns]))
```

```python
# outliers check

for i in numerical_columns:
  plt.boxplot(df[i])

  # Adding labels and title
  plt.xlabel(i)
  plt.ylabel('Frequency')
  plt.title(f'boxplot of {i}')
  plt.show()
```

```python
# Create a histogram to visualize the distribution of salaries
df['TotalPay'].hist()
plt.xlabel('Total Pay')
plt.ylabel('Number of Employees')
plt.title('Distribution of Salaries')
plt.show()

```

```python
# Create a scatter plot to visualize the relationship
plt.scatter(df['TotalPay'], df['Year'])
plt.xlabel('Total Pay')
plt.ylabel('Years of Service')
plt.title('Correlation between Total Pay and Years of Service')
plt.show()
```

```python
# Group the data by 'Agency' and calculate the mean salary for each group
agency_means = df.groupby('Agency')['TotalPay'].mean()
print(agency_means)
# Compare the average salaries across different Agency
plt.bar(agency_means.index, agency_means.values)
plt.xlabel('Agency')
plt.ylabel('Average Total Pay')
plt.title('Average Salaries by Agency')
plt.show()
```

```python
# Group the data by 'JobTitle' and calculate the mean salary for each group
JobTitle_means = df.groupby('JobTitle')['TotalPay'].mean()
print(JobTitle_means)
# Compare the average salaries across different JobTitles
plt.bar(JobTitle_means.index, agency_means.values)
plt.xlabel('JobTitle')
plt.ylabel('Average Total Pay')
plt.title('Average Salaries by JobTitle')
plt.show()
```

```python
# Group the data by 'Year' and calculate the mean salary for each group
Year_means = df.groupby('Year')['TotalPay'].mean()
print(Year_means)
# Compare the average salaries across different Year
plt.bar(Year_means.index, agency_means.values)
plt.xlabel('Year')
plt.ylabel('Average Total Pay')
plt.title('Average Salaries by Year')
plt.show()
```

```python
# Use a pie chart to represent the proportion of employees in different departments
department_counts = df['JobTitle'].value_counts()
top_departments = department_counts.head(20)
plt.figure(figsize=(30, 30))
plt.pie(top_departments, labels=top_departments.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Proportion of Employees in Each Department\n')
plt.show()
```

```python
department_counts = df['Agency'].value_counts()
plt.pie(department_counts, labels=department_counts.index, autopct='%1.1f%%')
plt.title('Proportion of Employees in Each Department')
plt.show()
```

```python
# Features with target

for i in numerical_columns:
  plt.figure(figsize=(8, 6))
  plt.scatter(df[i], df['TotalPay'], c='blue', marker='o')

  # Adding labels and title
  plt.xlabel(i)
  plt.ylabel('TotalPay')
  plt.title(f'Relation between the {i} and TotalPay')

  plt.show()
```

```python
# Calculate the correlation matrix
correlation_matrix = df.corr()

plt.figure(figsize=(10, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
```
